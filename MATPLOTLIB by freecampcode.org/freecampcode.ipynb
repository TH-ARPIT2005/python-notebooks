## Bar Chart

```python
#use of some plot functions will remain same in bar charts also
labels= ['A','B','C']
values = [1,4,2]

plt.bar(labels,values)
#here we take values and labels to make the bar graph
#we take values on y axis and labels on x axis

plt.figure(figsize=(6,4))

plt.show()
```

```python
labels= ['A','B','C']
values = [1,4,2]

bars=plt.bar(labels,values)

bars[0].set_hatch('/')#we can make different pattens using this function
bars[1].set_hatch('O')
bars[2].set_hatch('*')

plt.figure(figsize=(6,4))



plt.show()
```

```python
labels= ['A','B','C']
values = [1,4,2]

bars=plt.bar(labels,values)
#This is the another method of making patterns on the bars
patterns = ['/','O','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))



# bars[0].set_hatch('/')#we can make different pattens using this function
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')

plt.figure(figsize=(6,4))



plt.show()
```

```python
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[(fifa.Weight >= 200)].count()[0]
#here we divided different weight categories with differnt
#variable names and count() function will count how many weights
#lie in that catoegory

weights = [light,light_medium, medium , medium_heavy, heavy]
#we make the list of weights because we want to show the
#the weights on the pie chart

plt.pie(weights)
plt.show()
```

```python
labels = ['under 125','125-150','150-175','175-200','over 200']
#weigths will automatically take lables according to list in series or index number
plt.pie(weights , labels = labels)
plt.show()
```

```python
plt.style.use('default')

plt.figure(figsize = (5,8))

revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['FC Barcelona' , 'Real Madrid','NE Revolution']

boxes=plt.boxplot([barcelona,madrid,revs],labels=labels,patch_artist=True)#you can pass color argument here also but it will make everything of same color
#saving the result as boxes(variable here)

#this is logic for changing color for different part
for box in boxes['boxes']:
    #set edge color
    box.set(color='#4286f4',linewidth=2)#linewidth argument is to change line width

    #change fill color
    box.set(facecolor='#e0e0e0')#to change boxes color from inside or called face color
    #facecolor is another argument from documentation but this will show error so we have to pass argument in boxes that patch_artist = True


plt.title('Professional Soccer Team Comparison')
plt.ylabel("FIFA Overall Rating ")
plt.show()
```

```python
plt.style.use('default')

plt.figure(figsize = (5,8))

revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['FC Barcelona' , 'Real Madrid','NE Revolution']

boxes=plt.boxplot([barcelona,madrid,revs],labels=labels,patch_artist=True,medianprops={'linewidth':2})
#medianprops argument is used to change width of median showing line in graph or box
for box in boxes['boxes']:
    #set edge color
    box.set(color='#4286f4',linewidth=2)

    box.set(facecolor='#e0e0e0')


plt.title('Professional Soccer Team Comparison')
plt.ylabel("FIFA Overall Rating ")
plt.show()
```
