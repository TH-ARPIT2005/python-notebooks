### Pair Grid (in pair plot this is used to show particluar sets of data we want to see)

```python
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris,hue='species',)#first i want to create empty grid system based on this iris data
#after that i want to make scatter plot in upper lower and in dialonal
iris_g.map(plt.scatter)#it just make a scatter plot
```

```python
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris,hue='species',)
iris_g.map(plt.scatter)
iris_g.map_diag(plt.hist)#to create histogram diagonally
#we can make it more clear like in video we just need some 
#arguments from chatgpt
```

```python
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris,hue='species',)
#scatter plot removed
iris_g.map_diag(plt.hist)
iris_g.map_offdiag(plt.scatter)
#this will create scatter plot above and below the diagonal
#like map_diag function is used to create graph in diagonal
#same map_offdiag function is used to create the graphs off side of diagonal
```

```python
#but if we want different stuff on upper part and differnt stuff on our lower part
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris,hue='species',)
iris_g.map_diag(plt.hist)
iris_g.map_upper(plt.scatter)#this fuction is for creating in upper part of diagonal
iris_g.map_lower(sns.kdeplot)#this function is for creating in lower part of diagonal
#and also in recent version on seaborn we make kde plot like this sns.kdeplot
```

```python
iris = sns.load_dataset('iris')

iris_g = sns.PairGrid(iris, hue='species',
                      x_vars=['sepal_length','sepal_width'],
                      y_vars=['petal_length','petal_width'])
iris_g.map(plt.scatter)
iris_g.add_legend()
#we can make custom grid by defining x and y variables for 
#our custom grid in form of list , and above is the example
```

```python
#lets assign variables of different colors
tips_fg = sns.FacetGrid(tips_df, col='time', hue='smoker',height=4,aspect=1.3)
#here aspect is width based on height
tips_fg.map(plt.scatter,'total_bill','tip')
```

```python
#lets add some different options 
tips_fg = sns.FacetGrid(tips_df, col='time', hue='smoker',height=4,aspect=1.3,col_order=['Dinner','Lunch'],palette='Set1')
tips_fg.map(plt.scatter,'total_bill','tip',edgecolor='w')
#i made some changes in both lines , like col_order is use to condrol the order of category we are using
#and i also make the edgecolor = white
```

```python
#lets make some more changes like sizes , widths, lines,markers etc because this guy has a lots of options
kws = dict(s=50,linewidth = .5 , edgecolor ='w')
tips_fg = sns.FacetGrid(tips_df,col='sex',hue='smoker',height = 4 , aspect= 1.3,
                         hue_order=['Yes','No'],
                        hue_kws=dict(marker=['^','v']))
tips_fg.map(plt.scatter,'total_bill','tip',**kws)
#we did some crazy stuff right now , lets try to understand
#bro explain me every line in this code each and every argument , i want to understand this how is it working
#copy above line in chatgpt you will understand everything clearly 
```

```python
plt.figure(figsize=(8,6))
sns.set_context('paper',font_scale=1.4)
#now ploting regression plot
sns.lmplot(x='total_bill',y='tip',hue='sex',data=tips_df,markers=['o','^'],
           scatter_kws={'s':100, 'linewidths':0.5 , 'edgecolor':'w'})
#hue is use to show the seperation between men and women
#scatter_kws is use to make further changes in our scatter plot
#s = size for marker , and other details are according to there name
```

```python
#another regression plot
#what you can acctually do is that you can seperate the data 
#into seperate columns
sns.lmplot(x='total_bill',y='tip', col='sex',row='time',data=tips_df)
#Creates separate plots in different columns for each unique value in the 'sex' column ('Male', 'Female').
#Creates separate plots in different rows for each unique value in the 'time' column ('Lunch', 'Dinner').
```
